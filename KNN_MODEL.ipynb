```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
model = KNeighborsClassifier(n_neighbors=400)
```

```python
X = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
model.fit(X, y)
```

```python
new_data =[[32,150000]]
prediction = model.predict(new_data)
print("result :",prediction[0])
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X[y == 0]['Age'], X[y == 0]['EstimatedSalary'],
            color='red', label='Not Purchased (0)', alpha=0.6)
plt.scatter(X[y == 1]['Age'], X[y == 1]['EstimatedSalary'],
            color='green', label='Purchased (1)', alpha=0.6)
plt.scatter(new_data[0][0], new_data[0][1],
            color='blue', marker='*', s=200, label=f'New Data ({prediction[0]})')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('K-NN Classification: Age vs Estimated Salary')
plt.legend()
plt.grid(True)
plt.show()
```
